```python
times_naive, times_opti = naive_vs_sorted(naive_pareto, optimized_pareto)
np.savetxt('naive.csv', times_naive, delimiter=',')
np.savetxt('opti.csv', times_opti, delimiter=',')
```

```python
x = np.arange(200, 10001, 200)
plt.figure()
plt.plot(x, times_opti,label='algorithme optimal')
plt.plot(x, times_naive, label="algorithme naïf")
plt.title("Comparaison des temps de calcul des points Pareto non dominés")
plt.xlabel("n")
plt.ylabel('temps (s)')
plt.yscale('log')
plt.grid(True,which="both")
plt.legend()
plt.show()
```
